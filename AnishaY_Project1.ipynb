```python
import pandas as pd

# Raw URL for the CSV file
csv_url = 'https://raw.githubusercontent.com/aniyodta-sys/BDATA200/refs/heads/main/Data/AI%20Job%20Market%20Dataset.csv'

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(csv_url)

# Display the first 5 rows of the DataFrame
display(df.head())
```

### Data Dictionary

This data dictionary provides an overview of specific factors in the dataset, including their descriptive explanation and the count of non-null cells for each. This helps in understanding the relevance and completeness of these key data points.

```python
factors_description = {
    'job_id': 'Unique identifier for each job posting',
    'job_title': 'Role title such as Data Scientist, Machine Learning Engineer, Data Analyst, etc.',
    'company_size': 'Size category of the company',
    'company_industry': 'Industry sector of the hiring company',
    'country': 'Country where the job is located',
    'remote_type': 'Work arrangement (Remote, Hybrid, Onsite)',
    'experience_level': 'Level of experience required (Entry, Mid, Senior)',
    'years_experience': 'Number of years of experience required',
    'education_level': 'Minimum education qualification required',
    'skills_python': 'Whether Python skill is required (1 = Yes, 0 = No)'
}

new_data_dict_list = []

for col_name, description in factors_description.items():
    if col_name in df.columns:
        new_data_dict_list.append({
            'Factor': col_name,
            'Description': description,
            'Number of Cells': df[col_name].count() # Non-null count
        })

data_dictionary_df = pd.DataFrame(new_data_dict_list)
display(data_dictionary_df)
```

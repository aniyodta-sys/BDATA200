# Research Question:

Should we learn to code, learn to leverage AI to code, or learn to code AI? This project compares these three pathways in the modern AI-driven tech industry. The “learn to code AI” category focuses on AI/ML, data engineering, Deep Learning, Machine Learning, and SaaS-related roles to evaluate their job prospects, payoffs, and requirements, and whether they are “better” than the other pathways. It also examines whether there will still be demand for the other two categories—especially “learn to code” and “learn to leverage AI to code”—given that many computer science professionals are already shifting toward AI-assisted workflows and SaaS-based development becoming the standard.

### Why this matters?

This question matters because AI is no longer a niche skill in tech—it is becoming the standard. Recent labor market reports show that AI skills now appear in more than half of U.S. tech job postings, and employers increasingly treat AI fluency as a baseline expectation rather than a bonus. At the same time, software work is splitting into different paths: some professionals focus on strong coding fundamentals, others focus on leveraging AI tools and SaaS platforms to improve productivity, and others specialize in building AI itself through AI/ML, data engineering, Deep Learning, and Machine Learning roles.

These “learn to code AI” roles are among the fastest-growing and often offer stronger long-term demand and higher compensation, but they also require deeper specialization and a steeper learning curve. Meanwhile, many computer science professionals are already shifting into the second category—using AI to code—as tools like Copilot, LLM workflows, and SaaS-based development become standard practice. This creates an important question for data science students: where should we invest our limited time and effort? Understanding the tradeoff between learning difficulty, salary potential, and job security across these three paths can help students make smarter career decisions in an increasingly AI-driven industry.

### What I hope to discover:

I hope to discover a clear comparison between these three pathways in terms of required effort (learning curve), job opportunities, salary potential, and job security. Specifically, I want to understand the tradeoffs between each path so that data science and computer science students can make informed decisions about which skills to invest in during their limited learning period and how those decisions translate into real career outcomes.


# Methods (Preliminary Framework)
###1. Job Role Categorization
The dataset will be divided into three career pathways based on job titles and role function:

* Learn to Code: traditional software and coding-focused roles (e.g., Software Engineer, Data Analyst, Backend Developer)

* Learn to Leverage AI to Code: roles where coding is combined with AI tools, automation, and SaaS-driven workflows (interpreted based on job function and title context)

* Learn to Code AI: roles focused on building AI systems, including AI/ML, data engineering, Deep Learning, and Machine Learning positions (e.g., Machine Learning Engineer, Data Engineer, AI Engineer)

Since the dataset does not provide predefined categories, classification will be done manually based on job titles and role descriptions to ensure accuracy and meaningful grouping aligned with the research question.
###2. Payoff Variables (Career Benefits)

The following variables will be used to evaluate job “payoff” across the three categories:

* Company Size → proxy for organizational scale and prestige
* Company Industry → identifies where high-demand and high-value AI roles are concentrated
* Country → reflects geographic opportunity and potential relocation requirements
* Remote Type
* Salary
* Hiring_urgency → gives insight on fastest entry after graduation
* No. of Job Postings → gives insight on job security by analyzing how many jobs of that type are still hiring

###4. Tradeoff Variables (Required Investment / Effort)

The following variables will be used to measure the cost of entering each career pathway:

* Experience Level
* Years of Experience
* Education Level
* Skills (python, sql, ml, deep learning)

###5. Statistical and Analytical Methods

The analysis will use a combination of descriptive and inferential statistical approaches:

*   Descriptive statistics (mean, median, distribution analysis) will be used to summarize each category’s characteristics
*   Group comparisons of means will be used to compare job requirements across the three pathways
* Variance and standard deviation analysis will be used to measure consistency vs diversity of job requirements within each category
* The Central Limit Theorem (CLT) will justify comparing category means, as the large sample size ensures that the sampling distribution of means can be treated as approximately normal, making comparisons statistically reliable
* A composite effort–payoff index will be constructed using standardized variables (z-scores) to compare overall career “cost vs reward” across categories

## Group Comparison Analysis

### Section 1: Salary Comparison

**Understanding ANOVA and Tukey HSD:**

*   **One-Way ANOVA (Analysis of Variance):** ANOVA is a statistical test used to determine whether there are any statistically significant differences between the means of three or more independent (unrelated) groups. In this context, we use it to see if the average salaries across our three career pathways are significantly different from each other. The null hypothesis is that all group means are equal. A low p-value (typically < 0.05) indicates that we can reject the null hypothesis, suggesting at least one group mean is different.

*   **Tukey HSD (Honestly Significant Difference) Post-Hoc Test:** If ANOVA indicates a significant difference, it doesn't tell us *which* specific pairs of groups differ. Tukey HSD is a post-hoc test used after ANOVA to perform all pairwise comparisons between group means while controlling for the family-wise error rate. This means it helps us identify exactly which career pathways have statistically different average salaries.

```python
import scipy.stats as stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# 1. Salary Summary DataFrame
salary_comparison_summary = df.groupby('career_pathway')['salary'].agg(['mean', 'median', 'std'])
display(salary_comparison_summary.style.format("${:,.2f}"))

# Prepare data for ANOVA
salaries_learn_to_code = df[df['career_pathway'] == 'learn_to_code']['salary']
salaries_learn_to_code_ai = df[df['career_pathway'] == 'learn_to_code_ai']['salary']
salaries_leverage_ai_to_code = df[df['career_pathway'] == 'leverage_ai_to_code']['salary']

# 2. One-Way ANOVA Test
f_statistic_salary, p_value_salary = stats.f_oneway(
    salaries_learn_to_code,
    salaries_learn_to_code_ai,
    salaries_leverage_ai_to_code
)

print(f"\nOne-Way ANOVA (Salary):\nF-statistic = {f_statistic_salary:.2f}, p-value = {p_value_salary:.3f}")
if p_value_salary < 0.05:
    print("Interpretation: The differences in mean salary across the three career pathways are statistically significant at the 0.05 level.")
else:
    print("Interpretation: The differences in mean salary across the three career pathways are not statistically significant at the 0.05 level.")

# 3. Tukey HSD Post-Hoc Test
tukey_result_salary = pairwise_tukeyhsd(endog=df['salary'], groups=df['career_pathway'], alpha=0.05)
print("\nPairwise Tukey HSD Post-Hoc Test (Salary):")
display(tukey_result_salary)
```

### Section 2: Experience Comparison

**Understanding the Chi-square Test for Experience Level:**

The Chi-square ($\\chi^2$) test of independence is used to determine whether there is a significant association between two categorical variables. In this section, we apply it to `experience_level` (Entry, Mid, Senior) and `career_pathway` (Learn to Code, Leverage AI to Code, Learn to Code AI). The test helps us answer: "Is the distribution of experience levels across the career pathways significantly different?" In other words, are certain experience levels disproportionately represented in one pathway compared to others, or are the distributions similar? A low p-value (typically < 0.05) indicates that we can reject the null hypothesis of independence, suggesting a statistically significant association between career pathway and experience level.

```python
import pandas as pd
import scipy.stats as stats

# 1. Summary DataFrame for Years of Experience
years_experience_comparison_summary = df.groupby('career_pathway')['years_experience'].agg(['mean', 'std'])
display(years_experience_comparison_summary.style.format("{:.2f}"))

# Prepare data for ANOVA on years_experience
years_code = df[df['career_pathway'] == 'learn_to_code']['years_experience']
years_code_ai = df[df['career_pathway'] == 'learn_to_code_ai']['years_experience']
years_leverage = df[df['career_pathway'] == 'leverage_ai_to_code']['years_experience']

# 2. One-Way ANOVA Test on years_experience
f_statistic_years, p_value_years = stats.f_oneway(
    years_code,
    years_code_ai,
    years_leverage
)

print(f"\nOne-Way ANOVA (Years of Experience):\nF-statistic = {f_statistic_years:.2f}, p-value = {p_value_years:.3f}")
if p_value_years < 0.05:
    print("Interpretation: The differences in mean years of experience across the three career pathways are statistically significant at the 0.05 level.")
else:
    print("Interpretation: The differences in mean years of experience across the three career pathways are not statistically significant at the 0.05 level.")

# 3. Percentage breakdown table of experience_level
experience_level_pivot = df.groupby('career_pathway')['experience_level'].value_counts(normalize=True).mul(100).unstack(fill_value=0)
print("\nExperience Level Distribution per Pathway (Percentages):")
display(experience_level_pivot.style.format("{:.2f}%"))

# 4. Chi-square test on experience_level distribution
contingency_table = pd.crosstab(df['career_pathway'], df['experience_level'])
chi2_exp, p_value_exp, dof_exp, expected_exp = stats.chi2_contingency(contingency_table)

print(f"\nChi-square Test (Experience Level Distribution):\nChi2-statistic = {chi2_exp:.2f}, p-value = {p_value_exp:.3f}")
if p_value_exp < 0.05:
    print("Interpretation: There is a statistically significant association between career pathway and the distribution of experience levels at the 0.05 level.")
else:
    print("Interpretation: There is no statistically significant association between career pathway and the distribution of experience levels at the 0.05 level.")
```

### Section 3: Skill Demand Comparison

**Interpreting Skill Demand Comparison Results:**

This section uses the Chi-square test to examine if the demand for individual skills (Python, SQL, ML, Deep Learning, Cloud) is significantly different across the three career pathways. Each skill is treated separately. The results tell us:

*   **Skill-specific demand differences:** For each skill, if the p-value is less than 0.05, it indicates that the proportion of jobs requiring that skill is significantly different across the 'learn to code', 'leverage AI to code', and 'learn to code AI' pathways. This means the demand for that specific skill is *not* independent of the career pathway.

*   **Implications for the Research Question:**
    *   If a skill like 'Deep Learning' shows a significant difference, and its demand is highest in the 'Learn to Code AI' pathway, it strongly supports the idea that specializing in AI-building roles requires specific, advanced technical skills. This provides insight into the 'required investment/effort' for different paths.
    *   Conversely, if a skill like 'Python' shows no significant difference, it suggests that Python is a universally demanded skill across all three pathways, reinforcing its foundational importance regardless of specialization.

By comparing these demand patterns, students can make informed decisions about which skills to prioritize based on their chosen career path and the specific requirements of each domain within the AI-driven tech industry.

```python
import pandas as pd
import scipy.stats as stats

# 1. Summary DataFrame for Skill Demand Rates
skill_columns = ['skills_python', 'skills_sql', 'skills_ml', 'skills_deep_learning', 'skills_cloud']
skill_demand_rates_summary = df.groupby('career_pathway')[skill_columns].mean().mul(100)
skill_demand_rates_summary.columns = [col.replace('skills_', '') + '_demand_pct' for col in skill_demand_rates_summary.columns]
print("\nSkill Demand Rates per Pathway (Percentages):")
display(skill_demand_rates_summary.style.format("{:.2f}%"))

# 2. Chi-square test for each skill
chi_square_results = []

for skill in skill_columns:
    contingency_table_skill = pd.crosstab(df['career_pathway'], df[skill])
    chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table_skill)

    is_significant = 'Yes' if p_value < 0.05 else 'No'
    chi_square_results.append({
        'Skill': skill.replace('skills_', ''),
        'Chi2_Statistic': chi2,
        'P_Value': p_value,
        'Significant_at_0.05': is_significant
    })

# 3. Summarize chi-square results in a DataFrame
chi_square_df = pd.DataFrame(chi_square_results)
print("\nChi-square Test Results for Skill Demand Across Pathways:")
display(chi_square_df.style.format({
    'Chi2_Statistic': "{:.2f}",
    'P_Value': "{:.3f}"
}))

print("\nInterpretation for each skill:")
for index, row in chi_square_df.iterrows():
    skill_name = row['Skill']
    p_val = row['P_Value']
    if p_val < 0.05:
        print(f"- {skill_name}: The demand rates are statistically significant across career pathways (p = {p_val:.3f} < 0.05).")
    else:
        print(f"- {skill_name}: The demand rates are not statistically significant across career pathways (p = {p_val:.3f} >= 0.05).")
```

## Variance and Standard Deviation Analysis

### Section 1: Salary Consistency per Pathway

This section analyzes the consistency of salaries within each career pathway. While mean and median give us central tendencies, **variance** and **standard deviation** measure the spread or dispersion of salaries around that mean. A high standard deviation means salaries are widely spread, indicating more unpredictability.

To compare consistency fairly across groups with different mean salaries, we use the **Coefficient of Variation (CV)**. CV is calculated as (Standard Deviation / Mean) * 100. A **high CV** indicates that salary offers within that pathway are highly variable and unpredictable, meaning a wide range of salaries for similar roles. A **low CV** suggests that salaries are more consistent and reliable, making financial planning potentially easier for students. This analysis helps students understand the financial risk and predictability associated with each career path.

```python
salary_consistency = df.groupby('career_pathway')['salary'].agg(['var', 'std', 'mean'])
salary_consistency['CV'] = (salary_consistency['std'] / salary_consistency['mean']) * 100
display(salary_consistency.style.format({
    'var': "${:,.2f}",
    'std': "${:,.2f}",
    'mean': "${:,.2f}",
    'CV': "{:.2f}%"
}))

print("\nInterpretation (Salary Consistency):")
# Find the pathway with the lowest CV (most consistent salary)
most_consistent_salary_pathway = salary_consistency['CV'].idxmin()
min_cv_salary = salary_consistency['CV'].min()

# Find the pathway with the highest CV (most variable salary)
most_variable_salary_pathway = salary_consistency['CV'].idxmax()
max_cv_salary = salary_consistency['CV'].max()

print(f"- For a student seeking the most consistent and predictable salary outcomes, the '{most_consistent_salary_pathway}' pathway (CV: {min_cv_salary:.2f}%) seems to offer the most reliable compensation.")
print(f"- Conversely, the '{most_variable_salary_pathway}' pathway (CV: {max_cv_salary:.2f}%) indicates a wider spread of salaries, suggesting more variability and less predictability in potential earnings.")
print("This means that while 'learn_to_code_ai' offers higher average salaries, there might be a broader range of actual salaries experienced by individuals within that category.")
```

### Section 2: Experience Consistency per Pathway

This section examines the consistency of 'years_experience' required for each career pathway. Similar to salary, variance, standard deviation, and the Coefficient of Variation (CV) help us understand how predictable the experience requirements are. A lower CV in years of experience would mean that the expected experience level is more uniform across job postings within that pathway, making it easier for students to gauge the typical entry requirements.

```python
experience_consistency = df.groupby('career_pathway')['years_experience'].agg(['var', 'std', 'mean'])
experience_consistency['CV'] = (experience_consistency['std'] / experience_consistency['mean']) * 100
display(experience_consistency.style.format({
    'var': "{:.2f}",
    'std': "{:.2f}",
    'mean': "{:.2f}",
    'CV': "{:.2f}%"
}))

print("\nInterpretation (Experience Consistency):")
# Find the pathway with the lowest CV (most consistent experience requirement)
most_consistent_exp_pathway = experience_consistency['CV'].idxmin()
min_cv_exp = experience_consistency['CV'].min()

# Find the pathway with the highest CV (most variable experience requirement)
most_variable_exp_pathway = experience_consistency['CV'].idxmax()
max_cv_exp = experience_consistency['CV'].max()

print(f"- The '{most_consistent_exp_pathway}' pathway (CV: {min_cv_exp:.2f}%) shows the most consistent years of experience requirements, suggesting a more predictable expectation for job seekers.")
print(f"- In contrast, the '{most_variable_exp_pathway}' pathway (CV: {max_cv_exp:.2f}%) has more variable experience demands, meaning a wider range of years of experience could be sought by employers.")
print("Given that the mean years of experience were already found to be similar across pathways, consistent CVs further reinforce that experience predictability is fairly uniform.")
```

This points to a **pathway-specific skill**, which might be crucial for one specialization but less so for others. Understanding this helps students decide where to invest their learning efforts most effectively to align with their career goals.

```python
skill_columns = ['skills_python', 'skills_sql', 'skills_ml', 'skills_deep_learning', 'skills_cloud']

# Recalculate skill demand rates for each pathway (as percentages)
skill_demand_rates = df.groupby('career_pathway')[skill_columns].mean().mul(100)

# Calculate the standard deviation of these demand rates across the pathways for each skill
skill_demand_std = skill_demand_rates.std().to_frame(name='Std_Dev_Across_Pathways')
skill_demand_std.index = [col.replace('skills_', '') for col in skill_demand_std.index]

display(skill_demand_std.style.format("{:.2f}"))

print("\nInterpretation (Skill Demand Consistency Across Pathways):")
# Identify skills with lowest std dev (universally required)
universal_skills = skill_demand_std[skill_demand_std['Std_Dev_Across_Pathways'] < skill_demand_std['Std_Dev_Across_Pathways'].mean() / 2].index.tolist()
# Identify skills with highest std dev (pathway-specific)
pathway_specific_skills = skill_demand_std[skill_demand_std['Std_Dev_Across_Pathways'] >= skill_demand_std['Std_Dev_Across_Pathways'].mean() / 2].index.tolist()

print(f"- For a student deciding which skills to prioritize: Skills with low standard deviation across pathways, such as {', '.join(universal_skills) if universal_skills else 'None identified'}, are likely **universally required** skills. Investing in these provides a broad foundation valuable across all three pathways.")
print(f"- Skills with higher standard deviation across pathways, such as {', '.join(pathway_specific_skills) if pathway_specific_skills else 'None identified'}, may be more **pathway-specific**. Deepening expertise in these skills would be more beneficial for a particular specialization (e.g., Deep Learning for 'learn_to_code_ai').")
print("This means students can confidently invest in universally required skills knowing they'll be valuable in many paths, while strategically choosing pathway-specific skills for their desired specialization.")
```

### Section 4: Job Openings Consistency per Pathway

This section analyzes the consistency of 'job_openings' per pathway, which can serve as a proxy for job security consistency. A lower CV for job openings implies a more stable and predictable number of available positions within that career path, suggesting a more reliable job market for students entering that field.

```python
job_openings_consistency = df.groupby('career_pathway')['job_openings'].agg(['var', 'std', 'mean'])
job_openings_consistency['CV'] = (job_openings_consistency['std'] / job_openings_consistency['mean']) * 100
display(job_openings_consistency.style.format({
    'var': "{:.2f}",
    'std': "{:.2f}",
    'mean': "{:.2f}",
    'CV': "{:.2f}%"
}))

print("\nInterpretation (Job Openings Consistency):")
# Find the pathway with the lowest CV (most consistent job openings)
most_consistent_openings_pathway = job_openings_consistency['CV'].idxmin()
min_cv_openings = job_openings_consistency['CV'].min()

# Find the pathway with the highest CV (most variable job openings)
most_variable_openings_pathway = job_openings_consistency['CV'].idxmax()
max_cv_openings = job_openings_consistency['CV'].max()

print(f"- For students prioritizing a stable and predictable job market, the '{most_consistent_openings_pathway}' pathway (CV: {min_cv_openings:.2f}%) appears to offer the most consistent number of job openings.")
print(f"- The '{most_variable_openings_pathway}' pathway (CV: {max_cv_openings:.2f}%) suggests more variability in job opening numbers, which could imply less predictable job security.")
print("Overall, all pathways show relatively similar CVs for job openings, suggesting a generally consistent level of availability across the AI and data science job market.")
```

## Composite Effort-Payoff Index

This section constructs a Composite Effort-Payoff Index using z-score standardization to provide a holistic view of each career pathway. This index helps answer the research question by quantifying which pathway offers the best balance of high rewards (payoff) relative to the required investment (effort).

**How Z-Scores Work:**
A z-score measures how many standard deviations an element is from the mean. A positive z-score means the value is above the average, while a negative z-score means it's below. For this index, we're using z-scores to compare variables that have different units and scales (e.g., dollars for salary, years for experience) on a common, standardized scale.

*   **Payoff Variables (Higher is Better):** For salary, job openings, and hiring urgency score, a higher positive z-score means that pathway is better than average in that aspect. These scores directly contribute positively to the composite index.
*   **Effort Variables (Lower is Better):** For years of experience, total skills required, and education score, higher raw values mean more effort. To align these with our 'higher = better' index goal, we first calculate their z-scores and then *invert* them by multiplying by -1. So, a lower-than-average effort (e.g., fewer years of experience) will result in a positive inverted z-score, contributing positively to the composite index.

```python
from scipy.stats import zscore
from sklearn.preprocessing import MinMaxScaler

# Step 1 — Define your variables and create necessary columns/encodings
# Encode Hiring Urgency
hiring_urgency_mapping = {'Low': 1, 'Medium': 2, 'High': 3}
df['hiring_urgency_score'] = df['hiring_urgency'].map(hiring_urgency_mapping)

# Create Total Skills column
skill_columns = ['skills_python', 'skills_sql', 'skills_ml', 'skills_deep_learning', 'skills_cloud']
df['skills_total'] = df[skill_columns].sum(axis=1)

# Encode Education Level
education_mapping = {
    "High School": 1, # Assuming High School is lowest, will map to 1 if present
    "Bachelor's": 1,
    "Master's": 2,
    "PhD": 3
}
df['education_score'] = df['education_level'].map(education_mapping)

# Step 2 — Build a pathway-level summary DataFrame
composite_index_df = df.groupby('career_pathway').agg(
    salary=('salary', 'mean'),
    job_openings=('job_openings', 'mean'),
    hiring_urgency_score=('hiring_urgency_score', 'mean'),
    years_experience=('years_experience', 'mean'),
    skills_total=('skills_total', 'mean'),
    education_score=('education_score', 'mean')
)

# Step 3 — Standardize each of the six variables
# Use z-score for most variables
composite_index_df['salary_z'] = zscore(composite_index_df['salary'])
composite_index_df['job_openings_z'] = zscore(composite_index_df['job_openings'])
composite_index_df['hiring_urgency_score_z'] = zscore(composite_index_df['hiring_urgency_score'])
composite_index_df['years_experience_z'] = zscore(composite_index_df['years_experience'])
composite_index_df['skills_total_z'] = zscore(composite_index_df['skills_total'])

# Use Min-Max scaling for education_score to avoid precision loss due to very low variance
# Initialize MinMaxScaler
scaler = MinMaxScaler()
# Reshape data for scaler (expects 2D array)
composite_index_df['education_score_scaled'] = scaler.fit_transform(composite_index_df[['education_score']])
# For consistency with z-scores, we'll treat scaled values as 'z-scores' for the composite index
composite_index_df['education_score_z'] = composite_index_df['education_score_scaled'] * 2 - 1 # Scale to roughly -1 to 1 range, similar to z-scores for relative comparison

# Step 4 — Invert the effort scores
composite_index_df['years_experience_z_inverted'] = -1 * composite_index_df['years_experience_z']
composite_index_df['skills_total_z_inverted'] = -1 * composite_index_df['skills_total_z']
composite_index_df['education_score_z_inverted'] = -1 * composite_index_df['education_score_z']

# Step 5 — Compute the Composite Index
composite_index_df['composite_index'] = composite_index_df[[
    'salary_z', 'job_openings_z', 'hiring_urgency_score_z',
    'years_experience_z_inverted', 'skills_total_z_inverted', 'education_score_z_inverted'
]].mean(axis=1)

# Step 6 — Display results
final_display_df = composite_index_df[[
    'salary', 'job_openings', 'hiring_urgency_score',
    'years_experience', 'skills_total', 'education_score',
    'salary_z', 'job_openings_z', 'hiring_urgency_score_z',
    'years_experience_z_inverted', 'skills_total_z_inverted', 'education_score_z_inverted',
    'composite_index'
]].copy()

final_display_df = final_display_df.sort_values(by='composite_index', ascending=False)
final_display_df['rank'] = final_display_df['composite_index'].rank(ascending=False).astype(int)

print("\nComposite Effort-Payoff Index Results (Ranked):")
display(final_display_df.style.format({
    'salary': "${:,.0f}",
    'job_openings': "{:.2f}",
    'hiring_urgency_score': "{:.2f}",
    'years_experience': "{:.2f}",
    'skills_total': "{:.2f}",
    'education_score': "{:.2f}",
    'salary_z': "{:.2f}",
    'job_openings_z': "{:.2f}",
    'hiring_urgency_score_z': "{:.2f}",
    'years_experience_z_inverted': "{:.2f}",
    'skills_total_z_inverted': "{:.2f}",
    'education_score_z_inverted': "{:.2f}",
    'composite_index': "{:.2f}"
}))

# Step 7 — Add interpretation
print("\nInterpretation of Composite Effort-Payoff Index:")

best_pathway = final_display_df.index[0]
best_index = final_display_df['composite_index'].iloc[0]

print(f"- The pathway with the best effort-to-payoff ratio is '{best_pathway}' with a composite index of {best_index:.2f}.")
print(f"  This suggests that, when considering both rewards (salary, job openings, hiring urgency) and effort (years experience, total skills, education), the '{best_pathway}' pathway offers the most favorable balance for a student.")

print("\nLimitations of this index:")
print("- **Equal Weighting Assumption:** All six variables are currently weighted equally in the composite index. In reality, a student might prioritize salary much higher than, say, hiring urgency or education level. A more nuanced index would allow for custom weighting.")
print("- **Synthetic Dataset:** The dataset is synthetically generated, which, while useful for privacy and maintaining patterns, may not perfectly capture all intricacies and anomalies of the real-world job market.")
print("- **Limited Job Titles:** The analysis is based on only 6 distinct job titles, which might not fully represent the diverse landscape of AI and Data Science careers.")
print("- **Simplified Scoring:** Encoding for hiring urgency and education level is ordinal (1, 2, 3) and may not fully reflect the actual magnitude of differences between categories.")
```

### Visualization 3: Parallel Coordinates Plot - Career Pathway Profiles

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# --- Visualization 3: Radar Chart - Career Pathway Profiles ---

# Define mapping dictionaries for encoding categorical variables
education_mapping = {
    "Bachelor's": 1,
    "Master's": 2,
    "PhD": 3
}
experience_level_mapping = {'Entry': 1, 'Mid': 2, 'Senior': 3}
company_size_mapping = {'Startup': 1, 'Medium': 2, 'MNC': 3, 'Enterprise': 4}
hiring_urgency_mapping = {'Low': 1, 'Medium': 2, 'High': 3}
remote_type_mapping = {'Onsite': 1, 'Hybrid': 2, 'Remote': 3}

# Create encoded columns in a temporary DataFrame to avoid modifying the original df for this visualization
df_encoded = df.copy()
df_encoded['education_score'] = df_encoded['education_level'].map(education_mapping)
df_encoded['experience_score'] = df_encoded['experience_level'].map(experience_level_mapping)
df_encoded['company_size_score'] = df_encoded['company_size'].map(company_size_mapping)
df_encoded['hiring_urgency_score'] = df_encoded['hiring_urgency'].map(hiring_urgency_mapping)
df_encoded['remote_score'] = df_encoded['remote_type'].map(remote_type_mapping)

# Select variables for the radar chart and calculate their mean per pathway
radar_cols = [
    'education_score',
    'experience_score',
    'company_size_score',
    'hiring_urgency_score',
    'remote_score',
    'job_openings'
]

radar_data = df_encoded.groupby('career_pathway')[radar_cols].mean().loc[pathway_order]

# Normalize all variables to a 0-1 scale
scaler = MinMaxScaler()
normalized_radar_data = pd.DataFrame(scaler.fit_transform(radar_data), columns=radar_cols, index=pathway_order)

# Number of variables
num_vars = len(radar_cols)

# Calculate angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot is a circle, so we need to close the circle for each pathway
# This should be done for each pathway's data individually during plotting, not globally on the DataFrame columns
angles_closed = angles + [angles[0]]

# Axis labels
axis_labels = [
    'Education Level',
    'Experience Level',
    'Company Size',
    'Hiring Urgency',
    'Remote Type',
    'Job Openings'
]

# Plotting the Radar Chart
fig3, ax3 = plt.subplots(figsize=(9, 9), subplot_kw=dict(polar=True))

for i, pathway in enumerate(pathway_order):
    # Get the normalized data for the current pathway and close the polygon
    data_for_pathway = normalized_radar_data.loc[pathway].tolist()
    data_for_pathway_closed = data_for_pathway + [data_for_pathway[0]]

    ax3.plot(angles_closed, data_for_pathway_closed, color=palette[pathway], linewidth=2, label=pathway_labels[pathway])
    ax3.fill(angles_closed, data_for_pathway_closed, color=palette[pathway], alpha=0.25)

# Set axis limits and ticks
ax3.set_theta_offset(np.pi / 2)
ax3.set_theta_direction(-1)
ax3.set_rlabel_position(0) # Positions the r-labels (values) in the center of the first axis
ax3.set_xticks(angles)
ax3.set_xticklabels(axis_labels, fontsize=11)
ax3.set_ylim(0, 1)

# Add gridlines at 0.25, 0.50, 0.75, 1.0
ax3.set_yticks([0.25, 0.50, 0.75, 1.0])
ax3.set_yticklabels(['0.25', '0.50', '0.75', '1.0'], color="gray", size=9)
ax3.tick_params(axis='y', colors='darkgray')

ax3.set_title("Career Pathway Profiles Across Contextual Dimensions", fontsize=16, y=1.08) # Adjust title position
ax3.legend(title="Career Pathway", loc='upper right', bbox_to_anchor=(1.25, 1.1))

plt.figtext(0.5, 0.02, "Each polygon represents the mean value per pathway, normalized for comparability.", ha="center", fontsize=10, bbox={"facecolor":"white", "alpha":0.5, "pad":5})
plt.tight_layout(rect=[0, 0.05, 1, 1]) # Adjust rect for caption
plt.savefig('career_pathway_profiles_radar_chart.png', dpi=300)
plt.show()
```
